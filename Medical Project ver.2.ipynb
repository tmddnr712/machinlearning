```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#data csv 
data = pd.read_csv('insurance.csv')

#data column read
age = data['age']
sex = data['sex']
bmi = data['bmi']
children = data['children']
region = data['region']
charges = data['charges']


#label 
title = 'PREDICT THE DATA SET ON BASIC OF MALE AND FEMALE'

#age - charges
plt.scatter(data['age'], charges)
plt.xlabel('age')
plt.ylabel('charges')
plt.title(title)
line_fitter = LinearRegression()
line_fitter.fit(age.values.reshape(-1,1), charges)
plt.plot(age,line_fitter.predict(age.values.reshape(-1,1)))
plt.show()
print(line_fitter.intercept_)
print(line_fitter.predict([[25]]))

#sex - charges
plt.scatter(data['sex'], charges)
plt.xlabel('sex')
plt.ylabel('charges')
plt.title(title)
plt.show()

#bmi - charges
plt.scatter(data['bmi'], charges)
plt.xlabel('bmi')
plt.ylabel('charges')
plt.title(title)
line_fitter = LinearRegression()
line_fitter.fit(bmi.values.reshape(-1,1), charges)
plt.plot(bmi,line_fitter.predict(bmi.values.reshape(-1,1)))
plt.show()
print(line_fitter.intercept_)
print(line_fitter.predict([[20]]))

#children - charges
plt.scatter(data['children'], charges)
plt.xlabel('children')
plt.ylabel('charges')
plt.title(title)
line_fitter = LinearRegression()
line_fitter.fit(children.values.reshape(-1,1), charges)
plt.plot(children,line_fitter.predict(children.values.reshape(-1,1)))
plt.show()
print(line_fitter.intercept_)
print(line_fitter.predict([[0]]))


#smoker - charges
plt.scatter(data['smoker'], charges)
plt.xlabel('smoker')
plt.ylabel('charges')
plt.title(title)
plt.show()


#region - charges
plt.scatter(data['region'], charges)
plt.xlabel('region')
plt.ylabel('charges')
plt.title(title)
plt.show()










#histogram
#plt.hist(charges, bins=15)
#plt.show()
#plt.clf()
```
